```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randn(1000)
```

```python
plt.rcParams['font.sans-serif']=['SimHei']  #指定黑体
plt.rcParams['axes.unicode_minus']=False  #显示负号
```

```python
plt.figure(figsize=(10,8), dpi=300) # 设置分辨率
plt.hist(data)
plt.title('某某')
plt.xticks([-3,-2,-1,0,1,2,3], 
           ['2012-10-01','2012-10-02','2012-10-03','2012-10-04',
            '2012-10-05','2012-10-06','2012-10-07'],
           rotation=30)  # 刻度旋转30度
plt.savefig('./figure/myfig.jpg')
plt.show()  # 注意，要先保存，在show()，才能正常保存
```
