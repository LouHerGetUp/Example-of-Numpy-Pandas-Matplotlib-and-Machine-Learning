# 透视表（把表做变形）

```python
data = pd.pivot_table(ratings, index='user_id', columns='movie_id',values='rating')
```

# 为user_id为1的用户推荐电影

```python
data.info()
```

```python
data_center = data.apply(lambda x: x-x.mean(), axis=1)
```

```python
data_center
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_cos = []
for i in range(len(data)):
    sim_ = cosine_similarity(data.iloc[0].fillna(0).values.reshape(1,-1),
                             data.iloc[i].fillna(0).values.reshape(1,-1))
    sim_cos.append(sim_)
```

```python
sim_cos = [x[0][0] for x in sim_cos]
```

```python
data = data.assign(sim=sim_cos)
```

```python
data.head()
```

```python
data = data.sort_values(by='sim', ascending=False)
```

```python
data_top = data.iloc[1:6].copy() # 复制一份出来，后续操作不改变原数据
```

```python
data_top.dropna(how='all', axis=1, inplace=True)  # 删除5个人都没看过的电影，删除列
```

```python
data_top
```

```python
((~data_top.iloc[:,0].isnull()).astype(np.int32))*data_top.sim
# nan值变化为0，非nan变为1
```

```python
have_pre = data_top[data_top.columns[:-1]].apply(lambda x:((x*data_top.sim).sum())/(((~data_top.iloc[:,0].isnull()).astype(np.int32))*data_top.sim).sum())
```

```python
have_pre[:5]
```

```python
haven_see = data.columns[~data.iloc[0,:].isnull()][:-1]
```

```python
haven_see
```

```python
pre_movie_id = set(have_pre.index) - set(haven_see)
```

```python
pre_movie = have_pre[have_pre.index.isin(pre_movie_id)]
```

```python
recommend = pre_movie.sort_values(ascending=False)[:10]
```
