# 散点图

```python
x1 = [1,2,3,4,5]
y1 = [3,2,1,4,7]
x2 = [2,4,6,8,9]
y2 = [6,1,4,2,8]
```

```python
plt.scatter(x1,y1,color='r',marker='>', label='one', s=100)  # s=100 点的大小 
plt.scatter(x2,y2,color='b',marker='o', label='two', s=200)  # s可传动态参数
plt.legend()
```
