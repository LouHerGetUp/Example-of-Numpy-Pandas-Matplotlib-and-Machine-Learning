```python
import matplotlib.pyplot as plt
import numpy as np
```

# 堆积图

```python
comp=[1,2,3,4]
y1=[24,27,35,43]
y2=[30,35,17,12]
y3=[27,13,15,8]
y4=[19,25,33,37]
```

```python
labels = ['apple', 'micro', 'yahoo', 'google']
```

```python
plt.stackplot(comp, y1, y2, y3, y4, labels=labels)
plt.legend()
plt.xticks([1,2,3,4],['2011', '2012', '2013', '2014'])
plt.title('market')  # 占有率变化
```

# 饼图

```python
size = [15, 30 , 45, 10]
```

```python
explode = (0.1,0,0,0)  # 重点强调第一个
```

```python
plt.figure(figsize=(8,8))
plt.pie(size, labels=labels, autopct='%1.1f%%', explode=explode, startangle=90, shadow=True) 
# 1.1f%：域宽为1，小数点后为1，float，带百分号
# startangle 起始角度
```
