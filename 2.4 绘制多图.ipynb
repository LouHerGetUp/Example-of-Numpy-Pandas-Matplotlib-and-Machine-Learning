```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4]
y1 = [4,5,8,9]
y2 = [2,5,3,8]
```

```python
plt.subplot(2,1,1)
plt.plot(x, y1)
plt.subplot(2,1,2)
plt.plot(x, y2)
plt.suptitle('y1&y2')
```

```python
plt.subplot(2, 3, 1)
plt.plot(x, y1)
plt.subplot(2, 3, 2)
plt.plot(x, y1)
plt.subplot(2, 3, 3)
plt.barh(x, y1)
plt.subplot(2, 3, 4)
plt.scatter(x, y1)
plt.subplot(2, 3, 5)
plt.boxplot(x, y1)
```

```python
plt.subplot(2, 3, 1)
plt.plot(x, y1)
plt.subplot(2, 3, 2)
plt.plot(x, y1)
plt.subplot(2, 3, 3)
plt.barh(x, y1)
plt.subplot(2, 1, 2)
plt.plot(x, y2, 'b--')
```
