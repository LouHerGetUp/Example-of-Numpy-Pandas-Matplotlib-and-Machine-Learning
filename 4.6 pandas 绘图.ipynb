```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
import matplotlib
plt.style.use("ggplot")
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
# index为时间序列，持续时间为1000天
```

```python
ts
```

```python
ts = ts.cumsum()  # 做累加
```

```python
type(ts)
```

```python
ts.plot(title='sum')
plt.xlabel('time')
plt.ylabel('cum')
```

```python
df = pd.DataFrame(np.random.randn(1000,4), index=ts.index, columns=list('ABCD'))
```

```python
df = df.cumsum()  # 不做cumsum图像变化幅度会比较大
```

```python
df.plot()
```

```python
df['E'] = np.arange(1000)
```

```python
df.plot(x='E', y='A')  # 可以指定x，y轴
```

```python
df.iloc[5].plot.bar()
```

```python
df.iloc[5]
```

```python
df.iloc[5].plot.barh()
plt.axvline(0, color='r')  # 在x=0的地方画一条红色直线
```
