```python
import numpy as np
```

```python
data1 = np.random.randn(10).cumsum()   # randn正态分布的随机数,cumsum累加
data2 = np.random.randn(10).cumsum() 
```

```python
plt.plot(data1, label='data1')
plt.plot(data2, label='data2')
```

```python
x = np.linspace(0,7,100)  # 生成100个点，默认生成50个点
y = np.sin(x)
```

```python
x = [1,2,3,4]
y = [2,5,6,8]
y1 = [3,4,7,8]
```

```python
plt.plot(y, color='r')
plt.plot(y1, color='b')
```

```python
plt.plot(y, c='r', linestyle='--')
plt.plot(y1, c='b', linestyle='-.', linewidth=5)
```

```python
plt.plot(y, c='r', ls='--', lw=1, marker='o')
plt.plot(y1, c='b', ls='-.', lw=1, marker='^')  # marker标记
```

```python
plt.plot(y1, c='b', ls='-.', lw=1, marker='1')
plt.xlabel('age')
plt.ylabel('height')
plt.title('STUDENT',y=1.03) # y=1.03，标题位置上升
```

```python
y1 = np.random.randn(10).cumsum()
```

```python
plt.plot(y1, c='b', ls='-.', lw=1, marker='1',label='stu')
plt.legend(loc='best')
plt.xlabel('age')
plt.ylabel('height')
plt.title('STUDENT',y=1.03) # y=1.03，标题位置上升
plt.xlim(0,10) # x的取值范围
plt.ylim(0,5) # y的取值范围
plt.xticks([0,4,8],['born','baby','child'])  # 设置标记
plt.yticks([1,3,5],['height1','height2','height3'])  # 设置标记
plt.grid()  # 添加网格
```
