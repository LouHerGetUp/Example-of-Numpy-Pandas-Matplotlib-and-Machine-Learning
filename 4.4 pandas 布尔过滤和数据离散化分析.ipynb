```python
data = pd.DataFrame(np.random.randn(1000, 4))
```

```python
data[(np.abs(data) > 3).any(1)]  # 选取包含绝对值大于3的行 .any(1)->行
```

# 数据离散化分析

```python
data = pd.DataFrame(np.random.randint(1, 100, (50,2)))
```

```python
data
```

```python
bins = [20, 40, 60, 80, 100]
```

```python
cats  = pd.cut(data.iloc[:, 0], bins=bins)
```

```python
cats
```

```python
cats.value_counts()
```

```python
cats  = pd.cut(data.iloc[:, 0], bins=bins, right=False) 
# 左闭右开，默认左开右闭
```

```python
cats.value_counts()
```

```python
pd.cut(data.iloc[:,0], 4)   # 分成4类，从最小值到最大值区间平均划分
```

```python
pd.qcut(data.iloc[:,0], 4).value_counts()  # 分成4类，样本个数等分
```
